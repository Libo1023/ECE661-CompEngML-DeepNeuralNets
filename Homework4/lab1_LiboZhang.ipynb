# Problem (c)

```python
def count_nonzero (W) :
    count = 0
    for i in range(0, 5) :
        if W[i][0] != 0 :
            count = count + 1
    return count

def nonzero_min (W) :
    W1 = np.copy(W)
    W2 = np.absolute(W1)
    min_index = 0
    min_weight = W2[0][0] + W2[1][0] + W2[2][0] + W2[3][0] + W2[4][0]
    for i in range(0, 5) :
        if W2[i][0] == 0 :
            continue
        else :
            if W2[i][0] < min_weight :
                min_index = i
                min_weight = W2[i][0]
    return min_index

def prune_1 (W) :
    W1 = np.copy(W)
    min_index= nonzero_min(W1)
    W1[min_index][0] = 0
    return W1

def PGD (W) :
    W1 = np.copy(W)
    count = count_nonzero(W1)
    if count <= 2 :
        return W1
    else :
        if count == 3 :
            return prune_1(W1)
        if count == 4 :
            return prune_1(prune_1(W1))
        if count == 5 :
            return prune_1(prune_1(prune_1(W1)))
```

```python
W_init = np.zeros((5, 1))
r = 0.02
steps = 200
losses_c = list()
weights_c0 = list()
weights_c1 = list()
weights_c2 = list()
weights_c3 = list()
weights_c4 = list()
for i in range(0, steps) :
    for k in range(0, len(X)) :
        if i == 0 :
            if k == 0 :
                W_k = W_init
            else :
                W_k = W_k_plus_one
        else :
            W_k = W_k_plus_one
        
        d_k = Y[k][0]
        X_k = X[k].reshape((1, 5)).T
        s_k = np.dot(W_k.T, X_k)
        
        dL_dWk = (-2) * (d_k - s_k) * X_k
        
        #W_k_plus_one = W_k - r * dL_dWk
        project_weights = W_k - r * dL_dWk
        W_k_plus_one = PGD(project_weights)
        #print(W_k_plus_one)
        
    #W_k_plus_one = PGD(W_k_plus_one)
    losses_c.append(calculate_loss(Y, X, W_k_plus_one))
    weights_c0.append(W_k_plus_one[0][0])
    weights_c1.append(W_k_plus_one[1][0])
    weights_c2.append(W_k_plus_one[2][0])
    weights_c3.append(W_k_plus_one[3][0])
    weights_c4.append(W_k_plus_one[4][0])
    if i == steps - 1 :
        W_final_c = W_k_plus_one
        
#print(W_final_c)
```

# Problem (d)

```python
def calculate_L1_loss (Y, X, W, a) :
    loss = calculate_loss(Y, X, W)
    W_ori = np.array([W[0][0], W[1][0], W[2][0], W[3][0], W[4][0]])
    W_abs = np.absolute(W_ori)
    for i in range(0, len(W_abs)) :
        loss = loss + a * W_abs[i]
    return loss

def L1_GD (W, a) :
    W1 = np.copy(W)
    W_ori = np.array([W[0][0], W[1][0], W[2][0], W[3][0], W[4][0]])
    for i in range(0, len(W_ori)) :
        if W_ori[i] < 0 :
            W1[i][0] = (-1) * a
        if W_ori[i] == 0 :
            W1[i][0] = 0
        if W_ori[i] > 0 :
            W1[i][0] = a
    return W1

# lambda for L1
a = 0.2
#a = 0.5
#a = 1.0
#a = 2.0

W_init = np.zeros((5, 1))
r = 0.02
steps = 200
losses_d = list()
weights_d0 = list()
weights_d1 = list()
weights_d2 = list()
weights_d3 = list()
weights_d4 = list()
for i in range(0, steps) :
    for k in range(0, len(X)) :
        if i == 0 :
            if k == 0 :
                W_k = W_init
            else :
                W_k = W_k_plus_one
        else :
            W_k = W_k_plus_one
        
        d_k = Y[k][0]
        X_k = X[k].reshape((1, 5)).T
        s_k = np.dot(W_k.T, X_k)
        
        dL_dWk = (-2) * (d_k - s_k) * X_k + L1_GD(W_k, a)
        
        W_k_plus_one = W_k - r * dL_dWk
        
    losses_d.append(calculate_loss(Y, X, W_k_plus_one))
    weights_d0.append(W_k_plus_one[0][0])
    weights_d1.append(W_k_plus_one[1][0])
    weights_d2.append(W_k_plus_one[2][0])
    weights_d3.append(W_k_plus_one[3][0])
    weights_d4.append(W_k_plus_one[4][0])
    if i == steps - 1 :
        W_final_d = W_k_plus_one

#print(W_final_d)
```

# Problem (e)

```python
def proximal_operator (W, threshold) :
    W1 = np.copy(W)
    for i in range(0, 5) :
        if W[i][0] < ((-1) * threshold) :
            W1[i][0] = W[i][0] + threshold
        if W[i][0] > threshold :
            W1[i][0] = W[i][0] - threshold
        if W[i][0] >= ((-1) * threshold) and W[i][0] <= threshold :
            W1[i][0] = 0
    
    return W1


threshold = 0.004
#threshold = 0.01
#threshold = 0.02
#threshold = 0.04

W_init = np.zeros((5, 1))
r = 0.02
steps = 200
losses_e = list()
weights_e0 = list()
weights_e1 = list()
weights_e2 = list()
weights_e3 = list()
weights_e4 = list()
for i in range(0, steps) :
    for k in range(0, len(X)) :
        if i == 0 :
            if k == 0 :
                W_k = W_init
            else :
                W_k = W_k_plus_one
        else :
            W_k = W_k_plus_one
        
        d_k = Y[k][0]
        X_k = X[k].reshape((1, 5)).T
        s_k = np.dot(W_k.T, X_k)
        
        dL_dWk = (-2) * (d_k - s_k) * X_k
        #W_k_plus_one = W_k - r * dL_dWk
        proximal_operate = W_k - r * dL_dWk
        W_k_plus_one = proximal_operator(proximal_operate, threshold)
        
    losses_e.append(calculate_loss(Y, X, W_k_plus_one))
    weights_e0.append(W_k_plus_one[0][0])
    weights_e1.append(W_k_plus_one[1][0])
    weights_e2.append(W_k_plus_one[2][0])
    weights_e3.append(W_k_plus_one[3][0])
    weights_e4.append(W_k_plus_one[4][0])
    if i == steps - 1 :
        W_final_e = W_k_plus_one

#print(W_final_e)
```

# Problem (f)

```python
def gradient_L1 (value, threshold) :
    if value < ((-1) * threshold) :
        return (value + threshold)
    if value > threshold :
        return (value - threshold)
    if value >= ((-1) * threshold) and value <= threshold :
        return 0

def trimmed_gradient_L1 (W, threshold) :
    W1 = np.copy(W)
    W2 = np.absolute(W1)
    W3 = np.array([W2[0][0], W2[1][0], W2[2][0], W2[3][0], W2[4][0]])
    wmax = np.max(W3)
    min_index1 = np.argmin(W3)
    W3[min_index1] = wmax
    min_index2 = np.argmin(W3)
    W3[min_index2] = wmax
    min_index3 = np.argmin(W3)
    for i in range(0, 5) :
        if i == min_index1 or i == min_index2 or i == min_index3 :
            W1[i][0] = gradient_L1(W[i][0], threshold)
        else :
            W1[i][0] = W1[i][0]
    return W1

# lambda for L1
#a = 1.0
#a = 2.0
#a = 5.0
#a = 10.0

threshold = 0.02
#threshold = 0.04
#threshold = 0.1
#threshold = 0.2

W_init = np.zeros((5, 1))
r = 0.02
steps = 200
losses_f = list()
weights_f0 = list()
weights_f1 = list()
weights_f2 = list()
weights_f3 = list()
weights_f4 = list()
for i in range(0, steps) :
    for k in range(0, len(X)) :
        if i == 0 :
            if k == 0 :
                W_k = W_init
            else :
                W_k = W_k_plus_one
        else :
            W_k = W_k_plus_one
        
        d_k = Y[k][0]
        X_k = X[k].reshape((1, 5)).T
        s_k = np.dot(W_k.T, X_k)
        
        dL_dWk = (-2) * (d_k - s_k) * X_k # + trimmed_gradient_L1(W_k, a)
        #W_k_plus_one = W_k - r * dL_dWk
        to_trim = W_k - r * dL_dWk
        W_k_plus_one = trimmed_gradient_L1(to_trim, threshold)
        
    losses_f.append(calculate_loss(Y, X, W_k_plus_one))
    weights_f0.append(W_k_plus_one[0][0])
    weights_f1.append(W_k_plus_one[1][0])
    weights_f2.append(W_k_plus_one[2][0])
    weights_f3.append(W_k_plus_one[3][0])
    weights_f4.append(W_k_plus_one[4][0])
    if i == steps - 1 :
        W_final_f = W_k_plus_one

#print(W_final_f)
```
